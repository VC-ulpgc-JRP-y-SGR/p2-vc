Definida la imagen de salida de Canny, salida 0 o 255, obtiene el nmero de pxeles no nulos por columna y visualiza

```python

def show_histogram_x(image):
    #Si muestras el contenido de la imagen resultado, son valores 0 o 255
    #print(canny)
    #Cuenta el nmero de pxeles blancos (255) por columna
    #Suma los valores de los pixeles por columna
    col_counts = cv2.reduce(image, 0, cv2.REDUCE_SUM, dtype=cv2.CV_32SC1)

    #Normaliza en base al nmero de filas y al valor mximo del pxel (255)
    #El resultado ser el nmero de pxeles blancos por columna
    cols = col_counts[0] / (255 * image.shape[1])

    #Muestra dicha cuenta grficamente
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.axis("off")
    plt.title("Imagen")
    plt.imshow(image, cmap='gray') 

    plt.subplot(1, 2, 2)
    plt.title("Histograma vertical")
    plt.xlabel("Columnas")
    plt.ylabel("% pxeles")
    plt.plot(cols)
    #Rango en x definido por las columnas
    plt.xlim([0, image.shape[1]])

show_histogram_x(canny)
```

Webcam y sustraccin de modelo del fondo

```python
vid = cv2.VideoCapture(0)


# Fondo
# Inicializa la sustraccin del fondo con mezcla de gaussianas y deteccin de sombras
eliminadorFondo = cv2.createBackgroundSubtractorMOG2(history=100, varThreshold=50, detectShadows=True)
  


while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        # Aplica efecto espejo sobre la entrada
        framem=cv2.flip(frame, 1)

        
        #Con un segundo parmerto se puede definir mscara con zonas a actualizar
        objetos = eliminadorFondo.apply(framem)
        #objetos = eliminadorFondo.apply(framem, objetos, 0)  #No actualiza el fondo
        # Obtiene fondo
        background = eliminadorFondo.getBackgroundImage()

        # Muestra resultado
        cv2.imshow('Fotograma', objetos)
        # Muestra fondo
        cv2.imshow('Fondo', background)
  
   
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```
